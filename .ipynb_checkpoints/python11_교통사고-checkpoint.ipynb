# 교통사고 유형별 발생 분석
- 공공 데이터 교통사고 유형, 검색어: 전국교통사고다발지역
- csv 파일 다운

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# import matplotlib.dates as mdates

import pandas as pd
from datetime import datetime

d2_path = './ref/D2Coding-Ver1.3.2-20180524.ttf'
fm.fontManager.addfont(d2_path)
matplotlib.rcParams['font.family']='D2Coding'
```

```python
file_path = './ref/전국교통사고다발지역표준데이터.csv'
ori = pd.read_csv(file_path, encoding='EUC-KR')
ori
```

```python
ori_data = ori.copy()
ori_data = ori.groupby('사고유형구분')['사고건수'].sum().reset_index(name='count')
# ori_data.columns=['type','count']
ori_data = ori_data.sort_values('count', ascending=False)
ori_data = ori_data.rename(columns={'사고유형구분':'type'}) # 컬럼을  rename을 이용해서 하나만 바꿈, 두개를 넣고 싶으면 ,를 넣고 똑같이 적으면 된다 , '사고건수':'count'
ori_data
```

```python
ori_data.plot(kind= 'pie', 
              y = 'count', 
              labels=ori_data['type'], 
              figsize=(5,5), 
              startangle=0, 
              autopct='%2.1f%%', 
              title='각 사고 유형별 발생 통계',
              cmap=matplotlib.cm.Greens)
# ori_data.plot(kind= 'pie', y = 'count', labels=ori_data['type'])
plt.savefig('./chart/accident.png')
plt.show()
```

```python
ex_data = ori.copy()
ex_data= ex_data.rename(columns={'사고연도':'year','사고유형구분':'type','사고건수':'accident'})
ex_data
arr_data = list(set(ex_data['type']))
data_list=[]
for t in arr_data:
    data_list.append(ex_data[ex_data['type']== t])
print(arr_data,data_list)
```

```python
# import numpy as np

# ex_data = ori.copy()
# ex_data = ex_data.rename(columns={'사고유형구분': 'type'})

# unique_types = ex_data['type'].unique()
# arr_data = np.array(ex_data['type'])

# # 각 type별로 필터링하여 리스트 저장
# data_list = [ex_data[arr_data == t] for t in unique_types]
# print(data_list)
```

```python
total_list=[]
for d in data_list:
    result = d.groupby('year')["accident"].sum().reset_index(name='count')
    total_list.append(result)

# print(arr_data,total_list)

for i in range(len(arr_data)):    
    total_list[i].plot(
    kind = 'pie',
    y = 'count',
    labels = total_list[i]['year'],
    figsize=(7,7),
    startangle=90,
    autopct = '%.1f%%',
    title = f'연도별 {arr_data[i]} 사고 발생 통계',
    cmap = matplotlib.cm.Blues)
    # plt.show()
    plt.savefig(f'./chart/chart{i}.png')
```

주요 데이터 분석
1. 자전거 관련 사고가 가장 많음- 자전거 사고(24,939건)
최근 몇 년간 자전거 이용이 증가하면서 사고도 늘어나는 경향을 보이는 것으로 해석.
특히, 도심이나 자전거 도로 미비 지역에서 사고가 많을 가능성이 높음.

2. 보행자 관련 사고가 많음 (노인 + 어린이 포함)- 보행 노인 사고(23,165건)와 보행 어린이 사고(3,802건)를 합치면 약 2만 7천 건
여기에 스쿨존 어린이 사고(989건)까지 포함하면 보행 관련 사고가 자전거 사고보다 더 많아질 수도 있음.

3. 무단횡단 사고- (14,805건)
무단횡단을 줄이기 위한 시설 개선 (펜스, 신호등 개선) 및 계도 활동이 필요할 수 있음.


5. 스쿨존 어린이 사고는 상대적으로 적지만 주의가 필요
스쿨존에서 발생한 사고(989건)는 전체 중 가장 적지만, 어린이는 교통사고에서 특히 취약하므로 적은 숫자라도 큰 사회적 영향을 미침.


```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2,2, figsize=(10,8)) # figsize는 비율이라고 보면 됨, 2 * 2의 서브플롯 생성, 플롯은 차트가 나오는 것을 의미함

axs[0,0].plot([1,2,3],[4,5,6]) #,를 기준으로 앞은 x, 뒤는 y를 그림
axs[0,1].scatter([1,2,3],[7,8,1])
axs[1,0].bar(['A','B','C'],[4,7,5])
axs[1,1].hist([1,2,1,1,1,3,4,2,1,4],bins=10)

plt.tight_layout()
plt.show()
```
