```python
file_path = './ref/행정구역시도성연령별취업자_20250304.csv'
ori = pd.read_csv(file_path, encoding='EUC-KR')
ori.head()
```

```python
# 깊은 복사를 통해 원본에 손상이 없음
ex_data = ori.copy()
ex_data.columns = ['resion','year','age','gender','value']
ex_data.loc[ex_data['resion']=='세종특별자치시','value'] = ex_data.loc[ex_data['resion']=='세종특별자치시','value'].replace('-',0) # 세종특별자치시의 인구 수가  -로 표현되어 있어서 전처리 과정을 거침
ex_data
```

```python
ex_data['value'] = ex_data['value'].astype(int)
ex1_data = ex_data.groupby('year')['value'].sum().reset_index(name='total')
```

```python
ax = ex1_data.plot(x='year', y='total', figsize=(10,5), title='년도별 취업자 수')
ax.set_xticks(range(len(ex1_data)))
ax.set_xticklabels(ex1_data['year'],rotation=45)
plt.show()
```

```python
# def filter_gender(x):
#     return x['gender']=='남자'
# ex_data.filter(filter_gender)

m = ex_data.groupby('gender').get_group('남자')
wm = ex_data.groupby('gender').get_group('여자')
gender1 = pd.concat([m,wm], axis = 0)
gender1
```

연도별 취업자 수 통계 분석 보고서

1. 개요

2014년부터 2024년까지 연도 내 분기별 취업자 수 데이터를 분석한 결과이다. 분석 대상은 행정구역시도별 성별, 연령별 취업자 수의 변화추세다.

2. 연도별 현황 - 4분기마다 감소세 보임

3. 주요 변화 분석

1) 데이터 분석 결과, 2020년 취업률은 코로나19의 직접적인 영향을 받아 감소세를 나타냈다


4. 결론 및 전망
- 코로나19 이후 감소세를 보였지만 다시 상승하는 모양을 이루고 있음.
- 더 정확한 분석을 위해 나이별 차트가 필요해 보이기때문에 추후 추가 후 다시 작성하겠음



```python
ex3_data = ori.copy()
ex3_data.loc[:,'데이터']= ex3_data.loc[:,'데이터'].replace('-',0).astype(int)
# ex3_data.loc[:,'데이터']= ex3_data.loc[:,'데이터'].astype(int)
# ex3_data = ex3_data.fillna(0)
ex3_data
```

```python
filter0 = (ex3_data['시점']>='2018.1/4') & (ex3_data['시점'] <'2019.1/4')
ex3_data = ex3_data[filter0]
ex3_data.head()
```

```python
# ex3_data.loc[ex3_data['시도별']=='세종특별자치시','데이터'] = ex3_data.loc[ex3_data['시도별']=='세종특별자치시','데이터'].replace('-',0)
# ex3_data = ex3_data['데이터'].astype(int)
# ex3_data
```

```python
# filter1 = ex3_data['연령계층별'].isin(['20 - 29세'])
# filter2 = ex3_data['연령계층별'].isin(['60세이상'])
# ex3_data = ex3_data.loc['total'].astype(int)

# ex3_data = ex3_data.loc['데이터'].astype(int)

x_data = ex3_data[ex3_data['연령계층별'].isin(['20 - 29세'])]
# x_data.loc[:,'데이터'] = x_data.loc[:,'데이터'].astype(int)
y_data = ex3_data[ex3_data['연령계층별'].isin(['60세이상'])]
# y_data.loc[:,'데이터'] = y_data.loc[:,'데이터'].astype(int)
z_data = ex3_data
# z_data.loc[:,'데이터'] = ex3_data.loc[:,'데이터'].astype(int)


x = x_data.groupby('시도별')['데이터'].sum()#.reset_index(name='total') # 20대 취업자
y = y_data.groupby('시도별')['데이터'].sum()#.reset_index(name='total') # 60대 취업자
z = z_data.groupby('시도별')['데이터'].sum()#.reset_index(name='total') # 전체 취업자 수
z
```

```python
import numpy as np
regions = list(x.index) # 라벨
color = [np.random.rand(3,) for _ in range(len(regions))] # 랜덤 색상 color
# colors = list(color)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
z = np.array(z,dtype=float)
scatter = ax.scatter(x,y,s=z*0.3,color=color,alpha=0.5)

for n in range(len(regions)):
    ax.annotate(regions[n],xy=(x.iloc[n],y.iloc[n]), ha='center')

ax.plot([0,4000],[0,4000], color='b', linestyle='--', linewidth=2)

ax.set_title('2018년 전국 청년과 60세 이상 취업자 수 관계')
ax.set_xlabel('청년 취업자')
ax.set_ylabel('60세 이상 취업자')
plt.show()
```
