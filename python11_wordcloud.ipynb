```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.font_manager as fm
from konlpy.tag import Komoran #  

#   
font_path='./ref/D2Coding-Ver1.3.2-20180524.ttf'

#  
text = '''
SQL    SQL         .
   2022 7 NHN            ,    SQL         . 
        . 
 NHN 2013 2     (DB)  DB       DB  ,         .
   179386     13700  270    .
 SQL        ''  179386 .    , , ,   .
 ,  SQL         ,        (IP)    .    ,   .
  "         ,           " .
'''

# data = [t for t in text if bool(t.strip())]
# print(data)
#    
konoran = Komoran()
nouns = konoran.nouns(text)
nouns_text = ' '.join(nouns)
# print(nouns_text)

#  
wordcloud = WordCloud(font_path= font_path,width=800, height=400, background_color='lightgrey').generate(nouns_text)

# 
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')#  
plt.savefig('word2.png')
plt.show()
```
