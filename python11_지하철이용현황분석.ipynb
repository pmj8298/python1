```python
import pandas as pd
file_path = './ref/일별역별시간대별승하차인원(2025년 1월).csv'
df = pd.read_csv(file_path, encoding='EUC-KR')
df.head() #데이터 5개 가져오기(데이터 잘 가져오나 확인하는 용도)
# data.tail()
```

```python
columns0 = ['num', 'name', 'date', 'week', 'type', 'total']
time_col = [f'{i}-{i+1}h'for i in range(1,24)]
columns = columns0 + time_col + ['24-01h']
df.columns = columns
# data = data.iloc[:,:6]
df = df.loc[:,:'total']
```

```python
df.head()
```

```python
df.tail()
```

```python
data = df.groupby(['num', 'name', 'date']).sum().reset_index()
# data[data['date']=='2025-01-01']
data
```

```python
gdata= data[['name','date','total']]
gdata= gdata.sort_values(by=['name','date'], ascending=[True,True])
gdata.head(10)
```

```python
mdata = gdata.groupby(by='name').sum().reset_index()
tdata = mdata[['name','total']]
tdata
```

```python
sdata=tdata.sort_values(by='total', ascending=False)
sdata.head(10)
total_data = sdata[:20]
total_data
```

부산 지하철 이용현황 분석 보고서

1. 환승역의 영향: 부산 지하철에서 주요 환승역은 많은 이용객을 유치하는 경향이 보임.

- 서면역(1, 2호선 환승) → 부산 최대 상업 중심지, 유동 인구 많음.
- 연산역(1, 3호선 환승) → 환승으로 인한 이용객 증가.
- 사상역(2호선 및 경전철 환승) → 서부산 교통 중심지 역할, 버스터미널 존재.
- 동래역(1, 4호선 환승) → 교육 및 거주 지역, 환승 이용 증가.
- 덕천역(2, 3호선 환승) → 북부산 지역 주요 환승역.

2. 주요 상업·관광 지역의 영향
- 부산역(1,337,149명) → KTX와 연계, 여행객 유입.
- 센텀시티(1,083,780명) → 백화점 및 영화의 전당, 벡스코 등 유동인구 많음.
- 자갈치(1,070,458명), 남포(1,045,512명) → 관광지(자갈치시장, 국제시장) 영향.
- 해운대(803,371명), 장산(827,430명) → 관광 및 거주 지역 특성 반영.

4. 대학 및 행정기관 영향
- 부산대(760,441명) → 대학생 유입이 많음.
- 경성대·부경대(764,204명) → 대학가 특성 반영.
- 시청역(798,668명) → 공공기관 및 업무 지역 유동 인구 반영.

5. 결론 및 시사점
- 환승역을 중심으로 이용량이 높은 경향을 보임- 이는 환승 편의성과 교통 접근성이 중요한 요소임을 보여줌.
- 부산역, 센텀시티, 자갈치, 남포, 해운대 등 관광·상업 지역에서 높은 이용률을 보임. 이는 관광객 및 쇼핑객의 지하철 이용 비중이 크다는 점을 시사함.
