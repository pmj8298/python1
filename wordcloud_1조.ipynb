```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.font_manager as fm
from konlpy.tag import Komoran #  

# https://encykorea.aks.ac.kr/Article/E0029857

#   
font_path='./ref/D2Coding-Ver1.3.2-20180524.ttf'

#  
text = '''
 
  
      ,    .                . ,     ,       ,     .    ,     , () ,          .       .
          .   4     .  1414 ()          .               .          ()            .
     18   . ,  18  ()    .   (:  )  () () .      . , ()                   .           . ,       ,            .    .  18        .
     ,          . ()  ()  ()  ()        .        ,        .         .
,         .    ,      .      2 3   .          ()          .  ,        . , ()      .
        ()  ()    .              .             10 20   .      ,             .         ,        .
                .       ()   .      (:         ) ()   (:       )     .             .             ,     . ,             .
 
           .       ,     .     ,        . ,             10   .        .
'''

# data = [t for t in text if bool(t.strip())]
# print(data)
#    
konoran = Komoran()
nouns = konoran.nouns(text)
nouns_text = ' '.join(nouns)
# print(nouns_text)

#  
wordcloud = WordCloud(font_path= font_path,width=800, height=400, background_color='lightgrey').generate(nouns_text)

# 
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')#  
plt.savefig('word3.png')
plt.show()
```
